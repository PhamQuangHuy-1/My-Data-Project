```python
df = df.drop(df[df['Level'] == 0].index)
df = df.drop(index=0)
df = df.fillna(0)
```

- Level và Drop off by level tỷ lệ nghịch với nhau với correlation = -0.881685. Từ đây ta có thể thấy rằng khi chơi đến một mức level càng cao thì người chơi sẽ ít có xu hướng bỏ game hơn. 

- Win count tỷ lệ thuận rất chặt chẽ với số lượng drop off với correlation = 0.999304(~1). Từ đây ta có thể thấy rằng khi chơi thắng càng nhiều, thì người chơi sẽ có xu hướng không chơi nữa. Hay nói cách khác, độ khó của trò chơi ảnh hưởng rất nhiều đến khả năng từ bỏ game của người chơi. 

```python
level_nums = df['Level']
drop_offs = df['Drop off by level']
win_times = df['Win Play time (s)']
win_counts = df['Win_count']
lose_times = df['Lose Play time(s)']
lose_counts = df['Lose_count']

plt.figure(figsize=(12, 6))

# Drop-off 
plt.subplot(2, 3, 1)
plt.plot(level_nums, drop_offs, marker='o')
plt.xlabel('Level')
plt.ylabel('Drop-off Numbers')
plt.title('Drop-off Number by Level')

# Win Play Time
plt.subplot(2, 3, 5)
plt.plot(level_nums, win_times, marker='o')
plt.xlabel('Level')
plt.ylabel('Win Play Time (s)')
plt.title('Win Play Time by Level')

# Win Count
plt.subplot(2, 3, 2)
plt.plot(level_nums, win_counts, marker='o')
plt.xlabel('Level')
plt.ylabel('Win Count')
plt.title('Win Count by Level')

# Lose Play Time
plt.subplot(2, 3, 4)
plt.plot(level_nums, lose_times, marker='o')
plt.xlabel('Level')
plt.ylabel('Lose Play Time (s)')
plt.title('Lose Play Time by Level')

# Lose count
plt.subplot(2, 3, 3)
plt.plot(level_nums, lose_counts, marker='o')
plt.xlabel('Level')
plt.ylabel('Lose Count')
plt.title('Lose Count by Level')

plt.tight_layout()
plt.show()
```

```python
df = df[df['Level'] != 0]

level_nums = df['Level']
drop_offs = df['Drop off by level']
win_times = df['Win Play time (s)']
win_counts = df['Win_count']
lose_times = df['Lose Play time(s)']
lose_counts = df['Lose_count']

plt.figure(figsize=(12, 6))

# Win Play Time
plt.subplot(2, 2, 1)
plt.plot(drop_offs, win_times, marker='o')
plt.xlabel('Drop off Numbers')
plt.ylabel('Win Play Time (s)')
plt.title('Win Play Time by Drop off numbers')

# Win Count
plt.subplot(2, 2, 2)
plt.plot(drop_offs, win_counts, marker='o')
plt.xlabel('Drop off Numbers')
plt.ylabel('Win Count')
plt.title('Win Count by Drop off numbers')

# Lose Play Time
plt.subplot(2, 2, 3)
plt.plot(drop_offs, lose_times, marker='o')
plt.xlabel('Drop off Numbers')
plt.ylabel('Lose Play Time (s)')
plt.title('Lose Play Time by Drop off numbers')

# Lose count
plt.subplot(2, 2, 4)
plt.plot(drop_offs, lose_counts, marker='o')
plt.xlabel('Drop off Numbers')
plt.ylabel('Lose Count')
plt.title('Lose Count by Drop off numbers')

plt.tight_layout()
plt.show()
```

- Để chứng minh rõ hơn cho luận điểm trên, em sẽ tạo ra thêm 1 cột mới có tên là "Difficulty Rank" để xếp hạng độ khó theo các level sử dụng các yếu tố khác như Tỷ lệ drop off, số lần thua, số lần thắng, thời gian chơi thắng và thời gian chơi thua. 

```python
#df['Difficulty Rank'] = df['Drop off by level'] + df['Lose_count'] + (df['Lose Play time(s)'] / df['Win Play time (s)']) + (1 / df['Win_count'])
df['Difficulty Rank'] = df['Drop off by level'] + (df['Lose_count'] / df['Win_count']) + (df['Lose Play time(s)'] / df['Win Play time (s)'])
df['Difficulty Rank'] = df['Difficulty Rank'].rank(ascending=True)
df = df.sort_values(by='Difficulty Rank')
```

```python
df
```

```python
difficulty = df['Difficulty Rank']
plt.plot(level_nums,difficulty,marker="o")
plt.xlabel('Level')
plt.ylabel('Difficulty Rank')
plt.title('Comparison between Difficulty Rank and Level')
plt.show()
```

- Từ biểu đồ trên ta có thể thấy, Level và mức độ khó tỷ lệ thuận với nhau, hay nói cách khác level càng cao thì mức độ khó càng lớn.

```python
difficulty = df['Difficulty Rank']
plt.plot(drop_offs,difficulty,marker="o")
plt.xlabel('Drop off by level')
plt.ylabel('Difficulty Rank')
plt.title('Comparison between Difficulty Rank and Drop off by level')
plt.show()
```

- Từ biểu đồ trên ta có thể thấy rất rõ rằng tỷ lệ nghịch nhau của Drop off by Level và Difficulty Rank. Hay nói cách khác, khi độ khó của trò chơi càng cao thì người chơi sẽ ít có xu hướng bỏ game. Điều này có thể được lý giải là các level có độ khó thấp không hấp dẫn được người chơi, họ cảm thấy không đủ thử đủ thử thách nên đã bỏ game. 

- Tiếp theo chúng ta hãy phân tích chi tiết xem các mức độ nào cần điều chỉnh để cân bằng lại game, bằng cách tính sự chênh lệch giữa tỷ lệ thắng và tỷ lệ thua của người chơi qua từng level

```python
# Tính tỉ lệ thắng và tỉ lệ thua cho mỗi cấp độ
df['Win_rate'] = df['Win_count'] / (df['Win_count'] + df['Lose_count'])
df['Lose_rate'] = df['Lose_count'] / (df['Win_count'] + df['Lose_count'])

# Hiển thị biểu đồ tỉ lệ thắng/thua cho từng cấp độ
plt.plot(df['Level'], df['Win_rate'] - df['Lose_rate'], marker='o')
plt.xlabel('Level')
plt.ylabel('Win Rate - Lose Rate')
plt.title('Difference between Win Rate and Lose Rate across Levels')
plt.show()
```

```python
df['Win_rate - Lose_rate'] = df['Win_rate'] - df['Lose_rate']

# Sắp xếp DataFrame theo cột "Win_rate - Lose_rate" giảm dần
df_sorted = df.sort_values(by='Win_rate - Lose_rate', ascending=False)

# Lấy ra top 5 cấp độ có chênh lệch tỷ lệ thắng/thua lớn nhất
top_5_levels = df_sorted.head(5)['Level']
df_sorted
```
